```python
# IMPORTACIÓN DE LIBRERÍAS

from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *

from delta.tables import *

spark = SparkSession.builder.getOrCreate()
```

# Configuración de rutas

En esta sección se definen las rutas de los cuatro archivos CSV que serán utilizados durante el laboratorio.

Todos los archivos corresponden a la capa Bronze.

```python
path_estudiantes = "/Volumes/rendimiento_estudiantil/bronze/data/estudiantes.csv"
path_cursos = "/Volumes/rendimiento_estudiantil/bronze/data/cursos.csv"
path_matriculas = "/Volumes/rendimiento_estudiantil/bronze/data/matriculas.csv"
path_semestres = "/Volumes/rendimiento_estudiantil/bronze/data/semestres.csv"
```

# Carga de datos (Bronze)

Se cargan los cuatro archivos CSV respetando el esquema original.

Cada archivo constituye una fuente independiente dentro de la arquitectura Medallion.

```python
bronze_estudiantes = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv(path_estudiantes)
)

bronze_cursos = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv(path_cursos)
)

bronze_matriculas = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv(path_matriculas)
)

bronze_semestres = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv(path_semestres)
)
```

# Exploración inicial

Antes de comenzar el procesamiento de datos es importante verificar que cada dataset haya sido cargado correctamente.

```python
display(bronze_estudiantes)

display(bronze_cursos)

display(bronze_matriculas)

display(bronze_semestres)
```

```python
print("Cantidad de estudiantes:", bronze_estudiantes.count())
print("Cantidad de cursos:", bronze_cursos.count())
print("Cantidad de matrículas:", bronze_matriculas.count())
print("Cantidad de semestres:", bronze_semestres.count())
```

# Persistencia en Bronze

Una vez verificada la integridad de los archivos, cada DataFrame se almacena en formato Delta.

Esta capa conserva la información prácticamente sin transformaciones, permitiendo mantener una copia fiel de los datos originales.

```python
# Crear el esquema bronze si no existe
spark.sql("CREATE SCHEMA IF NOT EXISTS rendimiento_estudiantil.bronze")

(
    bronze_estudiantes.write
    .format("delta")
    .mode("overwrite")
    .saveAsTable("rendimiento_estudiantil.bronze.estudiantes")
)

(
    bronze_cursos.write
    .format("delta")
    .mode("overwrite")
    .saveAsTable("rendimiento_estudiantil.bronze.cursos")
)

(
    bronze_matriculas.write
    .format("delta")
    .mode("overwrite")
    .saveAsTable("rendimiento_estudiantil.bronze.matriculas")
)

(
    bronze_semestres.write
    .format("delta")
    .mode("overwrite")
    .saveAsTable("rendimiento_estudiantil.bronze.semestres")
)
```
